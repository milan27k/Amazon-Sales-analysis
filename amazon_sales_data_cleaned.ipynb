```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df['currency'] = df['currency'].fillna('INR')
```

```python
df['Amount'] = df['Amount'].fillna(0)
```

```python
df['fulfilled-by'].unique()
```

```python
df['fulfilled-by'] = df['fulfilled-by'].replace('Easy Ship','amazon')
df['fulfilled-by'] = df['fulfilled-by'].fillna('merchant')
```

```python
df['ship-city'] = df['ship-city'].fillna('unknown')
df['ship-state'] = df['ship-state'].fillna('unknown')
df['ship-country'] = df['ship-country'].replace('unknown','IN')
df['ship-country'] = df['ship-country'].replace('IN','INDIA')
df['ship-postal-code'] = df['ship-postal-code'].fillna('unknown')
```
